```python
# 1.2.2 正規分布のコード

# デフォルト引数は、標準正規分布
def normal(x, mu = 0, sigma = 1):
    y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu) ** 2 / (2 * sigma ** 2))
    return y
```

```python
# 正規分布の可視化

import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100) # -5から5までを100個に区切る
y = normal(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 平均を動かす
x = np.linspace(-10, 10, 1000)

y0 = normal(x, mu = -3)
y1 = normal(x, mu = 0)
y2 = normal(x, mu = 5)

plt.plot(x, y0, label = '$\mu$ = -3')
plt.plot(x, y1, label = '$\mu$ = 0')
plt.plot(x, y2, label = '$\mu$ = 5')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# 標準偏差を動かす
y0 = normal(x, mu = 0, sigma = 0.5)
y1 = normal(x, mu = 0, sigma = 1)
y2 = normal(x, mu = 0, sigma = 2)

plt.plot(x, y0, label = '$\sigma$ = 0.5')
plt.plot(x, y1, label = '$\sigma$ = 0.5')
plt.plot(x, y2, label = '$\sigma$ = 0.5')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# 中心極限定理の実験
N = 3

xs = []
for n in range(N):
    x = np.random.rand() # 0から1の間で、一様分布（すべての値が等しい確率で発生する分布）に従う乱数を生成
    xs.append(x)
    
x_mean = np.mean(xs)
print(x_mean) # ランダムな値の平均なので、毎回異なる値が出力される
```

```python
x_means = []
N = 10 # 1から10まで変えることで形状を見る

for _ in range(10000):
    xs = []
    for n in range(N):
        x = np.random.rand()
        xs.append(x)
    mean = np.mean(xs)
    x_means.append(mean)
    
# グラフの描画
plt.hist(
    x_means,
    bins = 'auto', # データに応じた適切な区間を設定
    density = True # 縦軸は確率密度（ヒストグラムの面積は1になる）
)
plt.title(f'N = {N}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```
